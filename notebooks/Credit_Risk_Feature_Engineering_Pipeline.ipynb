```python
loan_data = loan_data.drop_duplicates()
```

```python
loan_data.loc[loan_data['Experience'] > (loan_data['Age'] - 18), 'Experience'] = loan_data['Age'] - 18
loan_data
```

```python
loan_data['App_Month'] = loan_data['ApplicationDate'].dt.strftime('%Y-%m') # e.g., "2024-01"
loan_data['App_Quarter'] = loan_data['ApplicationDate'].dt.to_period('Q').astype(str)
loan_data
```

```python
loan_data['MonthlyObligations'] = loan_data['MonthlyDebtPayments'] + loan_data['MonthlyLoanPayment']
loan_data["MonthlyIncome"] = loan_data["AnnualIncome"]/12
loan_data['ResidualIncome'] = loan_data['MonthlyIncome'] - loan_data['MonthlyObligations']
loan_data[['MonthlyObligations','ResidualIncome','MonthlyIncome']]
```

```python
loan_data['Income_Stability'] = np.where(loan_data['JobTenure'] >= 2, 'Stable', 'Review Required')
loan_data[['EmploymentStatus','Income_Stability']]
```

```python
credit_bins = [0, 580, 670, 750, 900]
credit_labels = ['1. Poor', '2. Fair', '3. Good', '4. Very Good']
loan_data['Credit_Tier'] = pd.cut(loan_data['CreditScore'], bins=credit_bins, labels=credit_labels)
loan_data["Credit_Tier"].value_counts()
```

```python
def get_risk_band(score):
    if score <= 20: return 'Low Risk'
    elif score <= 40: return 'Medium Risk'
    elif score <= 60: return 'High Risk'
    else: return 'Critical'

loan_data['Risk_Band'] = loan_data['RiskScore'].apply(get_risk_band)
loan_data
```

```python
loan_data['DTI_Bucket'] = pd.cut(loan_data['DebtToIncomeRatio'],bins=[0, 0.3, 0.4, 0.5, 1.0], labels=['Healthy (<30%)', 'Moderate (30-40%)', 'High (40-50%)', 'Risky (>50%)'])
loan_data['DTI_Bucket']
```

```python
# Initialized with 'Normal'
loan_data['Risk_Segment'] = 'Normal'

# Bankruptcy or default history = immediate high risk
loan_data.loc[
    (loan_data['BankruptcyHistory'] == 1) |
    (loan_data['PreviousLoanDefaults'] == 1),
    'Risk_Segment'
] = 'High Risk (Adverse History)'

# DTI > 50% means half their income goes to debt before they buy food.
loan_data.loc[
    (loan_data['DebtToIncomeRatio'] > 0.50) &
    (loan_data['Risk_Segment'] == 'Normal'),
    'Risk_Segment'
] = 'High Risk (Over-Indebted)'

# Low DTI (<30%) and low utilization (<30%) and no inquiries
loan_data.loc[
    (loan_data['DebtToIncomeRatio'] < 0.30) &
    (loan_data['CreditCardUtilizationRate'] < 0.30) &
    (loan_data['NumberOfCreditInquiries'] == 0) &
    (loan_data['Risk_Segment'] == 'Normal'),
    'Risk_Segment'
] = 'Low Risk (Prime)'

print(loan_data['Risk_Segment'].value_counts())
```

```python
loan_data['ApplicationStatus'] = loan_data['LoanApproved'].map({1: 'Approved', 0: 'Rejected'})
loan_data[['EmploymentStatus','ApplicationStatus']]
```

```python
# IQR Method

cols_to_fix = ['AnnualIncome', 'LoanAmount', 'NetWorth', 'MonthlyIncome']

for col in cols_to_fix:
    # Calculating boundaries
    Q1 = loan_data[col].quantile(0.25)
    Q3 = loan_data[col].quantile(0.75)
    IQR = Q3 - Q1

    lower_limit = Q1 - 1.5 * IQR
    upper_limit = Q3 + 1.5 * IQR

    # TREATMENT IF NEEDED
    #loan_data[col] = loan_data[col].clip(lower=lower_limit, upper=upper_limit)

    outliers = loan_data[(loan_data[col] < lower_limit) | (loan_data[col] > upper_limit)]
    
    print(f"--- Outlier Report for {col} ---")
    print(f"Lower Bound: {lower_limit}")
    print(f"Upper Bound: {upper_limit}")
    print(f"Total Outliers: {len(outliers)}")
```

```python
bins = np.linspace(min(loan_data["AnnualIncome"]),max(loan_data["AnnualIncome"]),4)
labels = ["Low","Medium","High"]
loan_data["Income_Bracket"] = pd.cut(loan_data["AnnualIncome"], bins=bins, labels=labels, include_lowest=True)
loan_data["Income_Bracket"].value_counts()
```

```python
loan_data['AgeGroup'] = pd.cut(loan_data['Age'], bins=[17, 30, 45, 60, 100], labels=['Young', 'Middle-Aged', 'Senior', 'Retired'])
loan_data['AgeGroup'].value_counts()
```

```python
repay_counts = loan_data['PreviousLoanDefaults'].value_counts()
labels = ['Good Repayers (No Past Defaults)', 'Bad Repayers (History of Default)']
colors = ['#2ca02c', '#d62728']
plt.figure(figsize=(12, 5))

# Plot 1: The Pie Chart
plt.subplot(1, 2, 1)
plt.pie(repay_counts, labels=labels, autopct='%1.1f%%', colors=colors, startangle=90, explode=(0, 0.1))
plt.title('Repayment History of Applicants', fontsize=14)
```

```python
loan_data['Repayment_Status'] = loan_data['PreviousLoanDefaults'].map({0: 'Good Repayer', 1: 'Bad Repayer'})
palette = {'Good Repayer': '#2ca02c', 'Bad Repayer': '#d62728'}

plt.figure(figsize=(14, 10))

# --- Chart 1: Box Plot (Fixed) ---
plt.subplot(2, 2, 1)
# FIX: Added hue='Repayment_Status' and legend=False
sns.boxplot(x='Repayment_Status', y='RiskScore', data=loan_data,
            hue='Repayment_Status', legend=False, palette=palette)
plt.title('Risk Score Distribution: Good vs. Bad', fontsize=12)
plt.grid(True, alpha=0.3)
```

```python
risk_by_purpose = loan_data.groupby('LoanPurpose')['PreviousLoanDefaults'].mean().reset_index()
risk_by_purpose = risk_by_purpose.sort_values(by='PreviousLoanDefaults', ascending=False)

plt.subplot(2, 2, 4)
# FIX: For horizontal bars, y is the category, so we map hue=y
sns.barplot(x='PreviousLoanDefaults', y='LoanPurpose', data=risk_by_purpose,
            hue='LoanPurpose', legend=False, palette='mako')
plt.title('Default Rate by Loan Purpose', fontsize=12)
plt.xlabel('Bad Rate (%)')
plt.axvline(x=loan_data['PreviousLoanDefaults'].mean(), color='black', linestyle='--', label='Avg')

plt.tight_layout()
plt.show()
```

```python
Grouped_data = loan_data.groupby('LoanPurpose').agg({'LoanAmount': 'sum', 'LoanDuration': 'mean'})
labels = [f"{purpose}\n(Avg: {row['LoanDuration']:.1f} mo)"
    for purpose, row in Grouped_data.iterrows()]

plt.figure(figsize=(10, 7)) # Set the size of the chart

plt.pie(
    Grouped_data['LoanAmount'],   # The data determines slice size (Total Amount)
    labels=labels,                 # Our custom labels with duration info
    autopct='%1.1f%%',             # Show percentage of total amount
    startangle=140,                # Rotate start for better look
)

plt.title('Total Loan by Purpose\n(with Average Duration)', fontsize=14)
plt.axis('equal') # Equal aspect ratio ensures that pie is drawn as a circle.

# Show the plot
plt.show()
```

```python
grouped_data = loan_data.groupby('LoanPurpose').agg({'LoanAmount': 'sum', 'LoanDuration': 'mean'})
grand_total = grouped_data['LoanAmount'].sum()
labels = [f"{purpose}\n${row['LoanAmount']:,}\n(Avg: {row['LoanDuration']:.1f} mo)"
    for purpose, row in grouped_data.iterrows()]
number_of_slices = len(grouped_data)
colors = plt.cm.viridis(np.linspace(0, 1, number_of_slices))
plt.figure(figsize=(10, 8))
wedges, texts, autotexts = plt.pie(
    grouped_data['LoanAmount'],
    labels=labels,
    autopct='%1.1f%%',
    startangle=140,
    colors=colors,
    pctdistance=0.85, # Moves the % text into the ring
    wedgeprops={'width': 0.5, 'edgecolor': 'white'}
)
plt.setp(autotexts, size=9, weight="bold", color="white")
plt.text(
    0, 0,
    f"Total Loans\n${grand_total:,}",
    ha='center', va='center', fontsize=14, fontweight='bold', color='#333333'
)

plt.title('Loan Portfolio Distribution\n(Amount & Avg Duration)', fontsize=16)
plt.axis('equal')

plt.show()
```
